### This script is designed to automatically extract land tenure data from the SICAR public database. In the Sicar database, you have to manually download each land tenure for each municipality, so to speed up the process, I created this little bot. Please, bear in mind that at some point in the process of automaticaly scrape the data you have to enter some kind of captcha.

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
email='eduardominsky@hotmail.com' #Create a variable with my email in order to download data
driver = webdriver.Chrome('D:/Eduardo_Minsky/chromedriver')#Directory where your webdriver is located
driver.get("https://www.car.gov.br/publico/imoveis/index")#SICAR website


driver.find_element("link text", "BASE DE DOWNLOADS").click()
driver.find_element(By.XPATH, '//*[@id="base-downloads"]/div[1]/div/div/div[3]/button').click()
time.sleep(1.5)
driver.find_element(By.XPATH,"/html/body/div[1]/div/div[2]/div[2]/div/div/div[2]/div[27]/a").click()#Bear in mind that the last div[n] in this xpath, n represent the position of each brazilian state. So, if the xpath is like ".../div[1]/a" we are going to scrape data from the State of Acre, n = 2 is for Alagoas, and so on.
for i in range(999)# I put range of 999 because there is some states that have more than 800 municipalities, so it speed up the proccess:
    if i==0 :
        print('ignore it')
    elif i==1:
        print('ignore it')
    elif i>=2:
        print(i)
        driver.find_element(By.XPATH,'//*[@id="base-downloads"]/div[2]/div/div'+str([i])+'/div/div/button[1]').click()
        captcha = input("Please type the captcha: ")
        driver.find_element(By.XPATH,'//*[@id="form-email-download-base"]').send_keys(email)
        driver.find_element(By.XPATH,'//*[@id="form-captcha-download-base"]').send_keys(captcha)
        driver.find_element(By.XPATH,'//*[@id="btn-baixar-dados"]/span[1]').click()
        time.sleep(2.5)
        driver.find_element(By.XPATH,'//*[@id="modal-download-base"]/div/div/div[1]/button/span').click()
        time.sleep(2.5)
```
